# Chunking Strategy
A markdown file can be chunked based on its structure, such as headings, paragraphs, lists, and code blocks. This allows for better organization and readability. For example, you can create separate chunks for each section of the document, making it easier to navigate and understand the content.

````python
#get ast tree from markdown file
from marko import Markdown
md = Markdown()
with open("data/test_chunking.md", "r", encoding='utf-8') as f:
    text = f.read()
doc = md.parse(text)
````

Document is the root, and the sections are the branches. Each section can have its own sub-sections, creating a hierarchical structure that enhances the clarity and flow of the document. By using headings (e.g., # for main headings, ## for subheadings), you can easily identify and navigate through different parts of the content. Additionally, using lists and code blocks can further break down information into manageable chunks, making it easier for readers to digest and comprehend the material.

````python
#get content from a node
from marko.block import List
def extract_text(node):
    if not hasattr(node, 'children'):
        return ''
    
    if isinstance(node.children, str):
        return node.children
    
    if isinstance(node, List):
        return "\n".join(extract_text(child) for child in node.children)
    
    if isinstance(node.children, list):
        return "".join(extract_text(child) for child in node.children)
    
    return ''
````

````python
from marko.block import BlankLine, FencedCode, Heading, List, Paragraph, Quote, ThematicBreak
def read_md(doc):
    chunks = []
    current_chunk = {}
    for child in doc.children:
        if isinstance(child, BlankLine):
            continue
    
        if isinstance(child, ThematicBreak):
            continue
        
        if isinstance(child, Heading):
            if current_chunk:
                chunks.append(current_chunk)
            current_chunk = {
                'heading': extract_text(child),
                'content': ""
            }
            continue
        if isinstance(child, List):
            items = extract_text(child)
            current_chunk['content'] += ' ' + items
            continue
        
        if isinstance(child, Paragraph):
            content = extract_text(child)
            current_chunk['content'] += ' ' + content
            continue
        
        if isinstance(child, Quote):
            content = extract_text(child)
            current_chunk['content'] += ' ' + content
            continue
        
        if isinstance(child, FencedCode):
            code = extract_text(child)
            current_chunk['content'] += ' ' + code
            continue
    
    if current_chunk:
        chunks.append(current_chunk)
    
    return chunks
        
            
````
